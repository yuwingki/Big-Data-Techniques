## TASK 13. Counts of International flights 
### TASK 13a

```
inUSADF = flightsAllDF.filter((F.col('dest') == 'United States') & (F.col('origin') != 'United States')).groupby(['year']).sum('numflights').select('year', F.col('sum(numflights)').alias('totalflightsin')).sort('totalflightsin', ascending = True)
```

### TASK 13b

```
outUSADF = flightsAllDF.filter((F.col('dest') != 'United States') & (F.col('origin') == 'United States')).groupby('year').sum('numflights').select('year', F.col('sum(numflights)').alias('totalflightsout')).sort('totalflightsout', ascending = True)

inUSADF.join(outUSADF, on = 'year').select('*', (F.col('totalflightsin') > F.col('totalflightsout')).alias('morein')).sort('year').show()
```

## TASK 14. Aggregate statistics
### TASK 14a

```
flightsAllDF.filter((flightsAllDF.dest == 'United States') & ~(flightsAllDF.origin == 'United States')).groupby(['year']).agg(F.sum('numflights').alias('sumflightstoUS'), F.avg('numflights').alias('avgflightstoUS'), F.max('numflights').alias('maxflightstoUS')).sort('year', ascending=True).show()
```
