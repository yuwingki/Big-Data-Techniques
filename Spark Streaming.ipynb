## Task 12. Compute the averages of each column x, y, z and counts of Device column for each combination of values for Model and gt columns

```
#12a
gtmodelstatsQueryDF = streamingDF.groupby(['gt', 'Model']).agg(F.avg('x').alias('x_avg'), F.avg('y').alias('y_avg'), F.avg('z').alias('z_avg'), F.count('Device').alias('devicecount'))

gtmodelstatsQuery = gtmodelstatsQueryDF.writeStream.queryName('gtmodelstats').format("memory").outputMode("complete").start()
```

```
#12b
for x in range(0, 5):
    spark.sql("SELECT gt, Model, x_avg, y_avg, z_avg, devicecount FROM gtmodelstats").filter('gtmodelstats.x_avg > 0').show(5)
    sleep(5)
```

```
#12c
streamingQuery.stop()
```

## Task 13. Perform a join of a static DataFrame with a streaming DataFrame

```
#13a
historicalAvgDF = staticDF.groupby(['gt', 'Model']).agg(F.avg('x').alias('x_avg_h'), F.avg('y').alias('y_avg_h'), F.avg('z').alias('z_avg_h'))
```

```
#13b
streamingDF.groupby(['gt', 'Model']).agg(F.avg('x').alias('x_avg_s'), F.avg('y').alias('y_avg_s'), F.avg('z').alias('z_avg_s')).join(historicalAvgDF, on = (['gt', 'Model'])).writeStream.queryName('joindata').format("memory").outputMode("complete").start()
```

```
#13c
for x in range(0, 5):
    spark.sql("SELECT gt, Model, x_avg_h, y_avg_h, z_avg_h, x_avg_s, y_avg_s, z_avg_s FROM joindata").show(5)
    sleep(10)
```

```
#12c
streamingQuery.stop()
```
